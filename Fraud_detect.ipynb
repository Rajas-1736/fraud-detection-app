```python
corr = df.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
corr_target = corr['Class'].sort_values(ascending=False)

print(corr_target)
```

```python
X = df.drop('Class', axis=1)
y = df['Class']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Scale 'Amount' and 'Time'
X['Amount'] = scaler.fit_transform(X[['Amount']])
X['Time'] = scaler.fit_transform(X[['Time']])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

print("Training set:", X_train.shape)
print("Testing set:", X_test.shape)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print("Before SMOTE:", y_train.value_counts())
print("\nAfter SMOTE:", y_train_res.value_counts())
```

```python
print("Final Training Shape:", X_train_res.shape)
print("Final Testing Shape:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    roc_auc_score,
    roc_curve
)
```

```python
model = LogisticRegression(max_iter=1000)

model.fit(X_train_res, y_train_res)
```

```python
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

```python
roc_score = roc_auc_score(y_test, y_prob)
print("ROC-AUC Score:", roc_score)
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.plot(fpr, tpr, label="Logistic Regression")
plt.plot([0,1], [0,1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# Change threshold (try 0.3, 0.4, 0.6 etc.)
threshold = 0.3

y_pred_custom = (y_prob >= threshold).astype(int)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm_custom = confusion_matrix(y_test, y_pred_custom)

print("Confusion Matrix:\n", cm_custom)
print("\nClassification Report:\n", classification_report(y_test, y_pred_custom))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l2'],
    'solver': ['lbfgs']
}
```

```python
grid = GridSearchCV(
    LogisticRegression(max_iter=1000),
    param_grid,
    scoring='recall',   # IMPORTANT for fraud detection
    cv=3,
    n_jobs=-1
)

grid.fit(X_train_res, y_train_res)

print("Best Parameters:", grid.best_params_)
```

```python
grid = GridSearchCV(
    LogisticRegression(max_iter=1000),
    param_grid,
    scoring='recall',   # IMPORTANT for fraud detection
    cv=3,
    n_jobs=-1
)

grid.fit(X_train_res, y_train_res)

print("Best Parameters:", grid.best_params_)
```

```python
best_model = grid.best_estimator_

y_pred_best = best_model.predict(X_test)
y_prob_best = best_model.predict_proba(X_test)[:, 1]
```

```python
print(confusion_matrix(y_test, y_pred_best))
print(classification_report(y_test, y_pred_best))
```
